```python
import pandas as pd
import seaborn as sns
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv("Combined_News_DJIA.csv")
data.head(5)
```

```python
data.shape
```

```python
df_2 = data.copy().drop(['Label'],axis=1)
df_2.head(10)
```

```python
data = data.replace('b\"|b\'','',regex=True)
data.head(5)
```

```python
sia = SentimentIntensityAnalyzer()
sia.polarity_scores(" ")
```

```python
def detect_tb_subjectivity(text):
    return TextBlob(text).sentiment.subjectivity
def detect_tb_polarity(text):
    return TextBlob(text).sentiment.polarity
def detect_sia_pos(text):
    return sia.polarity_scores(text)['pos']

def detect_sia_neg(text):
    return sia.polarity_scores(text)['neg']

def detect_sia_comp(text):
    return sia.polarity_scores(text)['compound']
```

```python
cols = []
for i in range (1,26):
    col = ("Top{}".format(i))
    cols.append(col)

for col in cols:
    data[col] = data[col].astype(str)
    data[col+'_comp']=data[col].apply(lambda x:sia.polarity_scores(x)['compound'])
    data[col+'_tb']=data[col].apply(detect_tb_subjectivity)
    data[col+'tb_polarity'] = data[col].apply(detect_tb_polarity)
    data[col+'sia_pos'] = data[col].apply(detect_sia_pos)
    data[col+'sia_neg'] = data[col].apply(detect_sia_neg)
```
